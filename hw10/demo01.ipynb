```python
def make_env():
        env = make_retro(game='Airstriker-Genesis')
        env = wrap_deepmind_retro(env)
        return env
    
```

```python
venv = SubprocVecEnv([make_env] * 8)
ppo2.learn(
        network='cnn', 
        env=venv, 
        total_timesteps=int(100e6),
        nsteps=128,
        nminibatches=4,
        lam=0.95, 
        gamma=0.99, 
        noptepochs=4, 
        log_interval=1,
        ent_coef=.01,
        lr=lambda f : f * 2.5e-4,
        cliprange=0.1,
    )
```
